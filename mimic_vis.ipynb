# Figures

### Figure 1: Mortality Rate Bar Graphs

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
common_eth = list(admissions['ETHNICITY'].value_counts().index)[:6]
common_eth.remove('UNKNOWN/NOT SPECIFIED')
eth_df = admissions[admissions['ETHNICITY'].isin(common_eth)]
sns.barplot(data=eth_df, y='HOSPITAL_EXPIRE_FLAG', hue='ETHNICITY')
plt.title("Mortality Rate in Hospital by Ethnicity")
plt.ylabel("Mortality Rate")
plt.xlabel("Ethnicity")
plt.ylim(0, 0.2)

plt.subplot(1, 3, 2)
sns.barplot(data=admissions, y='HOSPITAL_EXPIRE_FLAG', hue='ADMISSION_TYPE')
plt.title("Mortality Rate in Hospital by Admission Type")
plt.ylabel("Mortality Rate")
plt.xlabel("Admission Type")
plt.ylim(0, 0.2)

plt.subplot(1, 3, 3)
sns.barplot(data=admissions, y='HOSPITAL_EXPIRE_FLAG', hue='INSURANCE')
plt.title("Mortality Rate in Hospital by Insurance Type")
plt.ylabel("Mortality Rate")
plt.xlabel("Insurance Type")
plt.ylim(0, 0.2)
```

### Figure 2: Ajustable Historgram Plotting Length of Stay for Each Age

```python
pat_stay = icustays.merge(patients, on='SUBJECT_ID')
pat_stay['AGE'] = pd.to_datetime(pat_stay['INTIME']).dt.year - pd.to_datetime(pat_stay['DOB']).dt.year
age_widget = widgets.IntSlider(min = 1, max = 100, description = 'Patient Age')

def plot_stay(age):
  plt.figure(figsize=(4, 4))
  age_df = pat_stay[(pat_stay['AGE']==age)]
  if age_df.any:
    sns.histplot(age_df['LOS'], binwidth=1, kde=True)
    plt.title(f'Histogram For Length of Stay for Patients Aged {age}')
    plt.xlim(0, 50)

widgets.interactive(plot_stay, age=age_widget)
```

### Figure 3: Time Series Displaying Hospital Occupancy Over the Course of the Year

```python
admissions['ADMITTIME'] = pd.to_datetime(admissions['ADMITTIME'])
admissions['DISCHTIME'] = pd.to_datetime(admissions['DISCHTIME'])
admissions['ADMITTIME'] = admissions['ADMITTIME'].apply(lambda x: x.replace(year=2000))
admissions['DISCHTIME'] = admissions['DISCHTIME'].apply(lambda x: x.replace(year=2000))

admissions.sort_values('ADMITTIME')

date_range = pd.date_range(start=admissions['ADMITTIME'].min(), end=admissions['DISCHTIME'].max(), freq='D')

count = pd.Series(0, index=date_range)

plt.figure(figsize=(12, 6))


for _, row in admissions.iterrows():
    mask = (date_range >= row['ADMITTIME']) & (date_range <= row['DISCHTIME'])
    count[mask] += 1

plt.figure(figsize=(12, 6))
count.plot()

plt.title('Hospital Occupancy Over Time')
plt.xlabel('Time')
plt.ylabel('Number of People in Hospital')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan']
plt.ylim(1300, 1800)

plt.show()
```

### Figure 4: Radial Bar Graph Displaying the Mortality Rates for the Top 7 Procedures With Non-Zero Mortality Rates

```python
proc_adm = procedures_icd.merge(admissions, on='SUBJECT_ID')
proc_adm['COUNTS'] = proc_adm.groupby('ICD9_CODE').size()

fig4 = (proc_adm[proc_adm['HOSPITAL_EXPIRE_FLAG'] == 1]
          .groupby('ICD9_CODE')
          .size()
          .div(proc_adm['COUNTS'])
          .multiply(100)
          .reset_index(name='Percentage')
          )

fig4['COUNT'] = proc_adm['COUNTS']
fig4 = fig4[fig4['Percentage'].notna()].sort_values('COUNT', ascending=False)
fig4 = fig4.rename({'index':'ICD9_CODE'}, axis='columns')
fig4 = fig4.merge(d_icd_procedures, on='ICD9_CODE')

df = fig4.head(7)
```

```python
ring_colours = ['#2f4b7c', '#665191', '#a05195','#d45087',
               '#f95d6a','#ff7c43','#ffa600']

ring_labels = [f' {v:.2f}%  {x}' for x, v in zip(list(df['LONG_TITLE']), list(df['Percentage']))]
data_len = len(df)

fig = plt.figure(figsize=(6,6), linewidth=6)

rect = [0.1,0.1,0.8,0.8]

ax_polar_bg = fig.add_axes(rect, polar=True, frameon=False)
ax_polar_bg.set_theta_zero_location('N')
ax_polar_bg.set_theta_direction(1)

for i in range(data_len):
    ax_polar_bg.barh(i, 100*1.5*np.pi/100, 
                     color='grey', 
                     alpha=0.1)
    
ax_polar_bg.axis('off')
    
ax_polar = fig.add_axes(rect, polar=True, frameon=False)
ax_polar.set_theta_zero_location('N')
ax_polar.set_theta_direction(1)
ax_polar.set_rgrids([0, 1, 2, 3, 4, 5, 6], 
                    labels=ring_labels, 
                    angle=0, 
                    fontsize=14, fontweight='bold', verticalalignment='center')

for i in range(data_len):
    ax_polar.barh(i, list(df['Percentage'])[i]*1.5*np.pi/100, 
                  color=ring_colours[i])


ax_polar.grid(False)
ax_polar.tick_params(axis='both', left=False, bottom=False, 
                   labelbottom=False, labelleft=True)
plt.title('Mortality Rate for Most Common Risky Procedures', fontsize=20, fontweight='bold', verticalalignment='center')
plt.show()

# I used https://www.andymcdonald.scot/create-eye-catching-radial-bar-charts-with-matplotlib as a reference for creating the radial chart
```

### Figure 5: Adjustible Plot Showcasing the Top Procedures for Each Common Diagnosis

```python
top_diag = diagnoses_icd['ICD9_CODE'].value_counts().head(10)
top_diag.index.tolist()
fig5 = diagnoses_icd.merge(procedures_icd, on='SUBJECT_ID')
fig5 = fig5[fig5['ICD9_CODE_x'].isin(top_diag.index)]
diagnoses_widget = widgets.Dropdown(options = top_diag.index.tolist(), description = 'Diagnosis:')
def diag_plot(diagnosis):
  plt.figure(figsize=(6, 6))
  top_procs = fig5[fig5['ICD9_CODE_x']==diagnosis]['ICD9_CODE_y'].value_counts().head(10)
  sns.barplot(y=[d_icd_procedures[d_icd_procedures['ICD9_CODE'] == i]['LONG_TITLE'].iloc[0] for i in top_procs.index], x=top_procs.values)
  plt.title(f"Top 10 Procedures for Subjects Diagnosed With {d_icd_diagnoses[d_icd_diagnoses['ICD9_CODE'] == diagnosis]['LONG_TITLE'].iloc[0]}")
  plt.xlabel("Count")
  plt.ylabel("Procedures")
widgets.interactive(diag_plot, diagnosis = diagnoses_widget)
```
